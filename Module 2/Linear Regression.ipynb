# Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

%matplotlib inline
```

## Generate Dummy Regression Data

```python
dataset = make_regression(n_features=1, n_samples=100, noise=5, random_state=42)
X_init = dataset[0]
y_init = dataset[1]
```

## Plotting the data

```python
plt.scatter(x=X_init, y=y_init)
```

```python
alpha = 0.1
iterations = 1500
```

```python
X = np.concatenate((np.ones((100,1)), X_init), axis=1)
print(X)
theta = np.zeros((2,1))
theta = np.matrix(theta)
X = np.matrix(X)
y=np.matrix(y_init)
print("y")
print(y)
X[0:5]
```

## Computing the cost J(Î¸)

```python
def computeCost(X, y, theta):
    m = len(X)
    cost = 0

    # ====================== YOUR CODE HERE ======================
    # Instructions: Compute the cost of a particular choice of theta
    res=np.dot(X,theta)
    #print(res)
    diff=np.subtract(res,y.T)
    cost=(np.square(diff).mean())/2
    
    #print(res)
    #print(y.T)
    
   
    return cost

computeCost(X, y, theta)
```

## Gradient descent

```python
def gradientDescent(X, y, theta, alpha, iterations):
    
    m = len(X) 
    res=np.dot(X,theta)
    diff=np.subtract(res,y.T)
    #print(np.dot(X_init.T,diff))
    newsum=(np.dot(X_init.T,diff)).sum()
    
    sumofdiff=diff.sum()
    theta[0][0]-=(alpha/m)*sumofdiff
    theta[1][0]-=(alpha/m)*newsum

    return theta
for i in range(iterations):
    theta = gradientDescent(X, y, theta, alpha, iterations)
    print(computeCost(X, y, theta))
```

## Plotting Linear Regression predictions

```python
plt.scatter(X_init, y_init, color='#003F72')
plt.plot(X_init, X*theta)
plt.show()
```
